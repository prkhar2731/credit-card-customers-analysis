```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/prakh/Desktop/credit-card-customers-analysis/data/Cleaned_bankchurnes.csv')
```

```python
print(df)
```

#Demographic Analysis



```python
print(df.columns)
```

```python
#Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Customer_Age'], kde=True, bins=30)
plt.title('Age Distribution of Credit Card Holders')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Income Category Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Income_Category', data=df)
plt.title('Income Category Distribution of Credit Card Holders')
plt.xlabel('Income Category')
plt.ylabel('Frequency')
plt.show()

```

```python
# Marital Status Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Marital_Status', data=df)
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

# Credit Usage Analysis


```python
# Salary (Income Category) vs. Credit Card Limit by Marital Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Income_Category', y='Credit_Limit', hue='Marital_Status', data=df)
plt.title('Income Category vs. Credit Card Limit by Marital Status')
plt.xlabel('Income Category')
plt.ylabel('Credit Limit')
plt.legend(title='Marital Status')
plt.show()

```

```python
#Age vs. Credit Card Limit by Card Category
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer_Age', y='Credit_Limit', hue='Card_Category', data=df)
plt.title('Age vs. Credit Card Limit by Card Category')
plt.xlabel('Age')
plt.ylabel('Credit Limit')
plt.legend(title='Card Category')
plt.show()
```

```python
# Age vs. Total Revolving Balance by Card Category
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer_Age', y='Total_Revolving_Bal', hue='Card_Category', data=df)
plt.title('Age vs. Total Revolving Balance by Card Category')
plt.xlabel('Age')
plt.ylabel('Total Revolving Balance')
plt.legend(title='Card Category')
plt.show()
```

```python
# Credit Card Limit by Card Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Card_Category', y='Credit_Limit', data=df)
plt.title('Credit Card Limit by Card Category')
plt.xlabel('Card Category')
plt.ylabel('Credit Limit')
plt.show()
```

Summary of Findings
Age Distribution: Understand the predominant age groups among the credit card holders.
Income Category Distribution: Provides insight into the income distribution of cardholders.
Marital Status Distribution: Proportion of different marital statuses among cardholders.
Interesting Patterns
Income Category vs. Credit Card Limit by Marital Status: Insights into how income and marital status influence credit limits.
Age vs. Credit Card Limit by Card Category: Differences in credit limits across various age groups and card categories.
Age vs. Total Revolving Balance by Card Category: Spending habits based on age and card type.
Credit Card Limit by Card Category: Different credit limits for different card categories.

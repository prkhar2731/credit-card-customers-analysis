```python
import pandas as pd

# Load the dataset
df = pd.read_csv('C:/Users/prakh/Desktop/credit-card-customers-analysis/data/BankChurners.csv')
# Display the first few rows
df.head()
```

```python
# Get summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

# Check data types
df.dtypes

# Get information about the dataset
df.info()

```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values

```

```python
from scipy.stats import zscore

# Compute Z-scores of the numerical columns
z_scores = df.select_dtypes(include=['float64', 'int64']).apply(zscore)

# Filter out outliers (keep rows with Z-score less than 3)
df_cleaned = df[(z_scores < 3).all(axis=1)]
```

```python
print("Original dataset shape:", df.shape)
print("Cleaned dataset shape:", df_cleaned.shape)
```

```python
# Example: Checking for negative values in columns where they don't make sense
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    if (df[column] < 0).any():
        print(f"Column {column} has negative values")

# Remove or correct erroneous data points as needed
# df = df[df['some_column'] >= 0]  # Example of removing rows with negative values in 'some_column'
```

### Data Preprocessing Decisions

1. **Handling Missing Values:**
   - Numerical values were filled with the mean of the respective columns.
   - Categorical values were filled with the mode of the respective columns.

2. **Handling Outliers:**
   - Outliers were identified using the Z-score method.
   - Rows with Z-scores greater than 3 in any numerical column were removed.

3. **Handling Erroneous Data Points:**
   - Checked for negative values in numerical columns and removed or corrected as needed.

These decisions ensure that the dataset is clean and ready for further analysis and modeling.
